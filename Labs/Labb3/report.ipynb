```python
import Labb3

myline = Labb3.line(-1, 0, "unlabelled_data.csv")
myline.display("blue")
line1 = Labb3.line(-0.489, 0, "unlabelled_data.csv")
line1.display("green")
line2 = Labb3.line(-2, 0.16, "unlabelled_data.csv")
line2.display("red")
line3 = Labb3.line(800, -120, "unlabelled_data.csv")
line3.display("yellow")

Labb3.plt.show()
print(f"Blue(Left/Right): {len(myline.left)}/{len(myline.right)}")
print(f"Green(Left/Right): {len(line1.left)}/{len(line1.right)}")
print(f"Red(Left/Right): {len(line2.left)}/{len(line2.right)}")
print(f"yellow(Left/Right): {len(line3.left)}/{len(line3.right)}")
```
